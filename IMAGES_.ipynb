```python
import os
import numpy as np
import pandas as pd
from PIL import Image
from tqdm import tqdm

# Set the path to your main directory containing sub-folders A to Z
path = '/content/drive/MyDrive/Colab Notebooks/UNZIP/New folder'

# Loop over each sub-folder
for folder_name in sorted(os.listdir(path)):
    folder_path = os.path.join(path, folder_name)
    if os.path.isdir(folder_path):  # Check if it is a directory
        images = []
        labels = []

        # Loop through each image in the sub-folder
        for image_file in tqdm(os.listdir(folder_path), desc=f'Processing {folder_name}'):
            file_path = os.path.join(folder_path, image_file)
            try:
                with Image.open(file_path) as img:
                    numpy_array = np.array(img)
                    flattened_array = numpy_array.flatten()
                    images.append(flattened_array)
                    labels.append(folder_name)
            except Exception as e:
                print(f"Failed to process {file_path}: {str(e)}")

        # Create a DataFrame for the current folder
        df = pd.DataFrame(images)
        df['label'] = labels  # Add the label column

        # Save this dataframe as a CSV file named after the folder
        csv_file_path = os.path.join("/content/drive/MyDrive/Colab Notebooks/csv_files", f"{folder_name}.csv")
        df.to_csv(csv_file_path, index=False)
        print(f"Data for {folder_name} saved to {csv_file_path}")
```

```python
df1=dff.sample(n=20000,random_state=42)
```

```python
df1
```

**Data Seggregation on sample dataset**

```python
y=df1["label"]
x=df1.drop(columns="label")
```

```python
from sklearn.model_selection import train_test_split as split
```

```python
x_train,x_test,y_train,y_test=split(x,y,train_size=0.75,random_state=42)
```

**Training**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
```

```python
#Logistic regression
LR_classifier = LogisticRegression()
LR_classifier.fit(x_train,y_train)
```

```python
y_test_pred_LR=LR_classifier.predict(x_test)
```

```python
y_test_pred_LR
```

```python
metrics.accuracy_score(y_test,y_test_pred_LR)
```

```python
#Decision tree
DT_classifier = DecisionTreeClassifier()

DT_classifier.fit(x_train,y_train)
```

```python
y_test_pred_DT=DT_classifier.predict(x_test)
```

```python
y_test_pred_DT
```

```python
metrics.accuracy_score(y_test,y_test_pred_DT)
```

```python
#svc
svm_classifier = SVC()

svm_classifier.fit(x_train,y_train)
```

```python
y_test_pred_svm=svm_classifier.predict(x_test)
```

```python
y_test_pred_svm
```

```python
metrics.accuracy_score(y_test,y_test_pred_svm)
```

```python
#RANDOMFOREST CLASSIFIER
RF_classifier = RandomForestClassifier()

RF_classifier.fit(x_train,y_train)
```

```python
y_test_pred_RF=RF_classifier.predict(x_test)
```

```python
y_test_pred_RF
```

```python
metrics.accuracy_score(y_test,y_test_pred_RF)
```

```python
#KNN CLASSIFIER
KNN_classifier = KNeighborsClassifier()

KNN_classifier.fit(x_train,y_train)
```

```python
y_test_pred_KNN=KNN_classifier.predict(x_test)
```

```python
y_test_pred_KNN
```

```python
metrics.accuracy_score(y_test,y_test_pred_KNN)
```
